```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_mousestudy_df = pd.merge(mouse_metadata, study_results,
                                  on='Mouse ID')
# Display the data table for preview
combined_mousestudy_df
```

```python
# Checking the number of mice.
combined_mousestudy_df["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
combined_mousestudy_duplicates_df = combined_mousestudy_df[combined_mousestudy_df.duplicated(subset=['Mouse ID','Timepoint'], keep=False)]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined_mousestudy_duplicates_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
combined_mousestudy_clean_df = combined_mousestudy_df.drop(combined_mousestudy_df[combined_mousestudy_df['Mouse ID'] == 'g989'].index, inplace = True)
combined_mousestudy_clean_df = combined_mousestudy_df.reset_index(drop=True)
combined_mousestudy_clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
combined_mousestudy_clean_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
Mean = combined_mousestudy_clean_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
Median = combined_mousestudy_clean_df.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
Variance = combined_mousestudy_clean_df.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
Standard_dev = Variance = combined_mousestudy_clean_df.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
SEM = combined_mousestudy_clean_df.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

drug_regimem_summary_df = pd.DataFrame({"Mean" : Mean,
                                       "Median" : Median,
                                       "Variance" : Variance,
                                       "Standard_Deviation" : Standard_dev,
                                       "SEM" : SEM})
drug_regimem_summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens: Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
maxtime = combined_mousestudy_clean_df.groupby("Mouse ID")["Timepoint"].max()
maxtime_df= pd.DataFrame(maxtime)
maxtime_df

#Merge dataframe to get total volume data
combined_maxtime_df = pd.merge(maxtime_df, combined_mousestudy_clean_df,
                                  on=['Mouse ID', 'Timepoint'])
combined_maxtime_df = combined_maxtime_df[combined_maxtime_df["Drug Regimen"].isin(['Capomulin','Ramicane', 'Infubinol', 'Ceftamin'])].reset_index(drop=True)
combined_maxtime_df
```

```python
# Create empty list to fill with tumor vol data (for plotting)
Capomulin = []
Ramicane = []
Infubinol = []
Ceftamin = []

#Run loop to append lists with tumor vol values
for x in range(len(combined_maxtime_df)):
    if combined_maxtime_df.iloc[x,2] == "Capomulin":
        Capomulin.append(combined_maxtime_df.iloc[x,6])
    if combined_maxtime_df.iloc[x,2] == "Ramicane":
        Ramicane.append(combined_maxtime_df.iloc[x,6])
    if combined_maxtime_df.iloc[x,2] == "Infubinol":
        Infubinol.append(combined_maxtime_df.iloc[x,6]) 
    if combined_maxtime_df.iloc[x,2] == "Ceftamin":
        Ceftamin.append(combined_maxtime_df.iloc[x,6])  
        
Capomulin_s = pd.Series(sorted(Capomulin))
Ramicane_s = pd.Series(sorted(Ramicane))
Infubinol_s = pd.Series(sorted(Infubinol))
Ceftamin_s = pd.Series(sorted(Ceftamin))
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 

#Capomulin
quartilesCapomulin = Capomulin_s.quantile([0.25,0.5,0.75])
lowerqCapomulin = quartilesCapomulin[0.25]
upperqCapomulin = quartilesCapomulin[0.75]
iqrCapomulin = upperqCapomulin-lowerqCapomulin
print(f"The lower quartile of Capomulin is: {lowerqCapomulin}")
print(f"The upper quartile of Capomulin is: {upperqCapomulin}")
print(f"The interquartile range of Capomulin is: {iqrCapomulin}")

# #Ramicane
quartilesRamicane = Ramicane_s.quantile([0.25,0.5,0.75])
lowerqRamicane = quartilesRamicane[0.25]
upperqRamicane = quartilesRamicane[0.75]
iqrRamicane = upperqRamicane-lowerqRamicane
print("-----------------------------------------------------------")
print(f"The lower quartile of Ramicane is: {lowerqRamicane}")
print(f"The upper quartile of Ramicane is: {upperqRamicane}")
print(f"The interquartile range of Ramicane is: {iqrRamicane}")

# #Infubinol
quartilesInfubinol = Infubinol_s.quantile([0.25,0.5,0.75])
lowerqInfubinol = quartilesInfubinol[0.25]
upperqInfubinol = quartilesInfubinol[0.75]
iqrInfubinol = upperqInfubinol-lowerqInfubinol
lower_boundInfubinol = lowerqInfubinol - (1.5*iqrInfubinol)
print("-----------------------------------------------------------")
print(f"The lower quartile of Infubinol is: {lowerqInfubinol}")
print(f"The upper quartile of Infubinol is: {upperqInfubinol}")
print(f"The interquartile range of Infubinol is: {iqrInfubinol}")
print(f"There is an outlier in Infubinol is that is lower than: {lower_boundInfubinol}")

#Ceftamin
quartilesCeftamin = Ceftamin_s.quantile([0.25,0.5,0.75])
quartilesCeftamin
lowerqCeftamin = quartilesCeftamin[0.25]
upperqCeftamin = quartilesCeftamin[0.75]
iqrCeftamin = upperqCeftamin-lowerqCeftamin
lower_boundCeftamin = lowerqCeftamin
upper_boundCeftamin = upperqCeftamin
print("-----------------------------------------------------------")
print(f"The lower quartile of Ceftamin is: {lowerqCeftamin}")
print(f"The upper quartile of Ceftamin is: {upperqCeftamin}")
print(f"The interquartile range of Ceftamin is: {iqrCeftamin}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig, (ax1, ax2, ax3,ax4) = plt.subplots(4, sharex = True)
fig.suptitle("Final Tumor Volume", fontsize=16, fontweight="bold")

ax1.set_xlim(combined_maxtime_df["Tumor Volume (mm3)"].min()-5, combined_maxtime_df["Tumor Volume (mm3)"].max()+5)
ax1.set_ylabel('Capomulin')
ax1.boxplot(Capomulin, vert=False)

ax2.set_ylabel('Ramicane')
ax2.boxplot(Ramicane, vert=False)

ax3.set_ylabel('Infubinol')
ax3.boxplot(Infubinol, vert=False)

ax4.set_ylabel('Ceftamin')
ax4.boxplot(Ceftamin, vert=False)
ax4.set_xlabel("Tumor Volume (mm3)")

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
Capomulin_df = combined_mousestudy_clean_df[combined_mousestudy_clean_df["Drug Regimen"] == "Capomulin"]
time = Capomulin_df[Capomulin_df["Mouse ID"] == "s185"]["Timepoint"]
tumvol = Capomulin_df[Capomulin_df["Mouse ID"] == "s185"]["Tumor Volume (mm3)"]
plt.plot(time, tumvol)
plt.title("Mouse ID: s185 Tumor Volume Over Time")
plt.xlabel("Time Point")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
tumvolavg = Capomulin_df.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
mouseweight = Capomulin_df.groupby("Mouse ID")["Weight (g)"].mean()

plt.scatter(mouseweight,tumvolavg)
plt.title("Tumor Volume vs Weight for Capomulin regimen")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

slope, intercept, r, p, std_err = st.linregress(
    mouseweight, tumvolavg)
regress_values = slope * mouseweight + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
                                                      
print(f"Linear Equation is : Tumor Volume (mm3) = {round(slope,2)} * Mouse Weight (g) + {round(intercept,2)} and r value is {round(r,2)}")

plt.scatter(mouseweight,tumvolavg)
plt.plot(mouseweight,regress_values,"r-")
plt.annotate(line_eq,(22,40),fontsize=15,color="red")
plt.title("Tumor Volume vs Weight for Capomulin regimen")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```
